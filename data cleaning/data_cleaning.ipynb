```python
import pandas as pd
```

```python
# Creamos un dataframe basado en el csv de clientes
df = pd.read_csv(r"C:\xampp\htdocs\EVA_3\clientes_banco.csv")
df
```

```python
# Vemos los datos nulos
print(df.isnull().sum())
```

```python
# Eliminamos datos duplicados si es que hay
df = df.drop_duplicates()
df
```

```python
# Hacemos que los id de los clientes esten ordenados en la base de datos c:
df = df.sort_values(by='Cliente_ID', ascending = True)
df
```

```python
# Revisamos cuantos valores nulos hay en la columna edad
print("Valores nulos en la columna edad:", df['Edad'].isnull().sum())
```

```python
# Rellenamos los valores nulos de edad con la mediana de edad
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
```

```python
# Ahora convertimos la edad a numero enteros
df['Edad'] = df['Edad'].astype(int)
# Vemos el resultado
df
```

```python
# Vemos cuantos valores nulos existen
print("Valores nulos en la columna genero:", df['Genero'].isnull().sum())
```

```python
# Reemplazamos los nulos con Desconocido
df['Genero'] = df['Genero'].fillna('Desconocido')
df
```

```python
# Nos aseguramos de normalizar los datos de genero, verificando que todos empiezen con mayuscula
df['Genero'] = df['Genero'].str.lower()
df['Genero'] = df['Genero'].replace({'femenino': 'Femenino', 'masculino': 'Masculino', 'desconocido': 'Desconocido'})
df
```

```python
# Vemos cuantos valores nulos existen
print("Valores nulos en la columna Saldo:", df['Saldo'].isnull().sum())
```

```python
# Rellenamos los valores nulos de saldo con 0
df['Saldo'] = df['Saldo'].fillna(0)
df
```

```python
# Ahora convertimos el saldo a numero enteros redondeando antes de convertir
df['Saldo'] = df['Saldo'].round().astype(int)
# Vemos el resultado
df
```

```python
# Cambiamos el indicador de cliente activo de 1 o 0 A Si y No
# Convretimos a string
df['Activo'] = df['Activo'].apply(lambda x: str(x))
# Reemplazamos 1.0 a si
df['Activo'] = df['Activo'].str.replace('1.0', 'Si')
# lo mismo de arriba
df['Activo'] = df['Activo'].str.replace('0.0', 'No')
# Adicionalmente, reemplazamos NaN con No, ya que si no tenemos datos de que el cliente siga activo, los mÃ¡s probable es que no este activo
df['Activo'] = df['Activo'].str.replace('nan', 'No')
# vemos los datos
df
```

```python
# Vemos cuantos valores nulos existen
print("Valores nulos en la columna Nivel_de_Satisfaccion:", df['Nivel_de_Satisfaccion'].isnull().sum())
```

```python
# Rellenamos los valores nulos de satisfaccion con la mediana
df['Nivel_de_Satisfaccion'] = df['Nivel_de_Satisfaccion'].fillna(df['Nivel_de_Satisfaccion'].median())
df
```

```python
# Ahora convertimos el nivel de satisfaccion a numero enteros redondeando antes de convertir
df['Nivel_de_Satisfaccion'] = df['Nivel_de_Satisfaccion'].round().astype(int)
# Vemos el resultado
df
```

```python
# Nuestros datos estan limpios :D, ahora podemos exportarlo a un csv
df.to_csv('datos_limpios.csv', index=False, encoding='utf-8')
```

```python
# Tambien podemos generar dos csv que separen los usuarios activos y no activos

# Primero separamos ambos tipos de cliente
usuarios_activos = df[df['Activo'] == 'Si']
usuarios_inactivos = df[df['Activo'] == 'No']

# Tambien podemos ver cuantos usuarios de cada tipo existen
print(f"Usuarios activos: {len(usuarios_activos)} registros")
print(f"Usuarios inactivos: {len(usuarios_inactivos)} registros")

# Y exportamos ambos csv
usuarios_activos.to_csv('usuarios_activos.csv', index=False, encoding='utf-8')
usuarios_inactivos.to_csv('usuarios_inactivos.csv', index=False, encoding='utf-8')
```
